```python
import numpy as np
```

### Generando un Vector Aleatorio de 12 números (del 1 al 10) enteros con Random

```python
vector = np.random.randint(1, 11, 12) #Tercer parámetro indica la cantidad de elementos
vector
```

### Obtención de la Dimensión de un Arreglo

```python
np.shape(vector)
```

### Transformar el vector de 12 elementos en una Matriz
El método reshape() redimensiona una estructura



* Con estructura de 4x3





```python
matriz = vector.reshape(4, 3)
matriz
```

* Con estructura de 6x2

```python
matriz2 = vector.reshape(6, 2)
matriz2
```

### Consultando la Dimensión, pero de la Matriz

```python
np.shape(matriz)
```

### Conociendo la Biblioteca Matplotlib
El paquete Matplotlib permite crear visualizaciones (gráficos) para interpretar resultados y data

```python
import matplotlib.pyplot as plt
```

### Creando un Gráfico Simple

```python
plt.plot(vector, label = 'Vector')
plt.title('Gráfico del Vector')
plt.xlabel('Indice (Posición)') #Título del eje x
plt.ylabel('Valores') #Título del eje y

plt.grid(False) #Agrega una grilla al gráfico
plt.legend() #Muestra una etiqueta, que se obtiene de la propiedad label del método plot()

plt.show() #Permite visualizar el gráfico
```

### Creando dos vectores aleatorios en el mismo gráfico

```python
vector1 = np.random.randint(1,11, 12)

vector2 = np.random.randint(5, 21, 12)

plt.plot(vector1, marker = 's', label = 'Vector 1') #la propiedad marker agrega un marcador de los valores del vector
plt.plot(vector2, marker = 'o', label = 'Vector 2')

plt.title('Comparando dos Vectores')
plt.xlabel('Indice (Posición)') #Título del eje x
plt.ylabel('Valores') #Título del eje y

plt.grid(True) #Agrega una grilla al gráfico
plt.legend() #Muestra una etiqueta, que se obtiene de la propiedad label del método plot()

plt.show() #Permite visualizar el gráfico

```

### Creando un Gráfico de Barra

```python
plt.bar(range(len(vector2)), vector2) #Método bar permite plotear un gráfico en barra
```
